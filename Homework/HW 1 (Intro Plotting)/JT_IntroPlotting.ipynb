# Problem 1

Generate a list of one hundred data points $x_n$ that are equally spaced in the interval $[0,2\pi]$. Use this to evaluate
$f(x_n) = \exp[6 \sin(x_n)]$, and then produce a plot of $x_n$ vs. $f(x_n)$. 

Produce two versions of the plot, one where
the data are represented by red diamonds that are disconnected, and a second, where the data is not shown as
markers, but instead a continuous green curve. In both graphs, label the axes and title the plots "Problem 1A" and "Problem 1B."

```python
x_list = np.linspace(0, 2*np.pi, 100)

def f(x):
    '''
    Evaluate and return f(x) = e^(6sin(x))
    '''
    func = np.exp(6*np.sin(x))
    return func

f_list = f(x_list)
```

```python
plt.scatter(x_list, f_list, color="red", marker="d", s = 20)

plt.xlabel(r"$x_n$")
plt.ylabel(r"$f(x) = \exp(6\sin(x_n))$")
plt.title("Problem 1A")

plt.show()
```

```python
plt.plot(x_list, f_list, color="green")

plt.xlabel(r"$x_n$")
plt.ylabel(r"$f(x) = \exp(6\sin(x_n))$")
plt.title("Problem 1B")

plt.show()
```

Histograms are like bar charts but each thing on the $x$-axis is a numerical range. We're going to make the ranges be the IGN scores, grouping them as intervals $[4,5)$, $[5,6)$, etc up to $[9,10]$ (only the last bin is right-inclusive). Specific bins chosen wrt actual set of data. On the $y$-axis will be the amount of consoles whose fighting game average is in that range.

Our second plot will be a line graph, showing the trend in RPG scores throughout time on consoles of three different companies: Nintendo, Sony, and Microsoft.

## Plot 1 (Histogram) Math

```python
fighting_data = df['Fighting'].to_numpy()
print("Data:\n", fighting_data)

bin_min = np.floor(np.min(fighting_data))
bin_max = np.ceil(np.max(fighting_data))
bins = np.arange(bin_min, bin_max + 1, 1)
bins
```

## Plot 2 (Line Plot) Math

```python
nin_list =  ["Nintendo 64", "GameCube", "Wii", "x"]
sony_list = ["PlayStation" , "PlayStation 2", "PlayStation 3", 'PlayStation 4']
ms_list =   ["x", "Xbox", "Xbox 360", "Xbox One"]
```

```python
print(df['Platform'])
```

```python
def find_console_indices(console_names, df):
    '''
    Find the indices of a given set of consoles in the dataframe. Takes a list of strings of console names
    and a dataframe, returns a list where each entry is the index of the corresponding console in the dataframe.
    '''
    console_idx = np.array([])
    platforms = np.array(df['Platform'])
    for console in console_names:
        console_where = np.where(console == platforms)
        console_idx = np.append(console_idx, console_where)

    return console_idx

nin_idx =  find_console_indices(nin_list, df)
sony_idx = find_console_indices(sony_list, df)
ms_idx =   find_console_indices(ms_list, df)
```

```python
nin_RPG =  np.array(df['RPG'][nin_idx])
sony_RPG = np.array(df['RPG'][sony_idx])
ms_RPG =   np.array(df['RPG'][ms_idx])
```

## Actual Plotting

```python
fig, ax = plt.subplots(2,1, figsize=(7,12))

fig.suptitle("Problem 4", fontsize=18)

ax[0].set_title("Average IGN Score for Fighting Games across several platforms.")
ax[0].set_xlabel("Average IGN Score of Fighting Games")
ax[0].set_ylabel("No. of Consoles")
ax[0].hist(fighting_data, bins)

marker_size = 10
ax[1].set_title("IGN Score for RPG Games through console generations.")
ax[1].set_xticks([1,2,3,4], labels=['Fifth Gen', 'Sixth Gen', 'Seventh Gen', 'Eighth Gen'])

ax[1].plot([1,2,3], nin_RPG, color = 'red', marker = 'x', markersize=marker_size, label='Nintendo')
ax[1].plot([1,2,3,4], sony_RPG, color = 'blue', marker = 'o', markersize=marker_size, label='Sony')
ax[1].plot([2,3,4], ms_RPG, color = 'green', marker = '^', markersize=marker_size, label='Microsoft')
ax[1].legend()

fig.show()
```

Note that there was no Fifth Gen Microsoft console (they got into the market with the first Xbox), and the dataset contained data for neither of the Eighth generation Nintendo consoles (WiiU or Switch).
